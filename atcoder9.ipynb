```python
N, M = map(int, input().split())
used = set()    
count = 0    
for _ in range(M):
    r, c = map(int, input().split())
        
    p1 = (r, c)
    p2 = (r, c + 1)
    p3 = (r + 1, c)
    p4 = (r + 1, c + 1)
        
    if (p1 not in used) and (p2 not in used) and (p3 not in used) and (p4 not in used):
        count += 1
        used.add(p1)
        used.add(p2)
        used.add(p3)
        used.add(p4)
            
print(count)
```
