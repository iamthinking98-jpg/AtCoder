```python
N, M = map(int, input().split())
n = list([True] * N for _ in range(N))
count = 0
for i in range(M):
    r, c = map(int, input().split())
    r -= 1
    c -= 1
    if n[r][c] == True and n[r][c+1] == True and n[r+1][c] == True and n[r+1][c+1] == True:
        n[r][c] = False
        n[r+1][c] = False
        n[r][c+1] = False
        n[r+1][c+1] = False
        count += 1
print(count)
```
