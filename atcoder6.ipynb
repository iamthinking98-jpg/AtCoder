```python
X,Y,Z = map(int, input().split())
A = (X - Z*Y) / (Z - 1)
a = int(A)
if A == a and a >= 0:
    print("Yes")
else:
    print("No")
```

```python
N = int(input())
A_list = list(map(int, input().split()))
T = 102
print(-1)
for i in range(1, N):
    for j in range(i):
        if A_list[i] < A_list[j]:
            T = j+1
    if T == 102:
        print(-1)
    else:
        print(T)
    T = 102
```

```python
S = input()
r = 0
count = 0
N = len(S)
for l in range(N):
    for r in range(l+1, N, 2):
        A = S[l:r+1]
        M =  int((r - l + 1)/2)
        B = sorted(A)
        if A == "".join(B) and len(set(A)) == 2 and int(B[M-1])+1 == int(B[M]):
            count += 1
print(count)
```
