```python
N,M = map(int, input().split())
A = list(map(int, input().split()))
judge = False
for i in range(N):
    B = A.copy()
    B.pop(i)
    if sum(B) == M:
        judge = True
        break
if judge:
    print("Yes")
else:
    print("No")
```
