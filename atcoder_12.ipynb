```python
N = int(input())
T = list(map(int, input().split()))
dict_T = {index: value for index, value in enumerate(T)}
dict_T = dict(sorted(dict_T.items(), key=lambda x: x[1]))
key_T = list(dict_T.keys())
A = []
for i in range(3):
    A.append(str(key_T[i]+1))

print(" ".join(A))    
```
