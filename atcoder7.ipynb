```python
import sys
input = sys.stdin.readline

def solve():
    N, M = map(int, input().split())
    counts = [0] * (M + 1)
    totals = [0] * (M + 1)

    for _ in range(N):
        category, value = map(int, input().split())
        counts[category] += 1
        totals[category] += value

    for j in range(1, M + 1):
        if counts[j] == 0:
            print(0) 
        else:
            avg = totals[j] / counts[j]
            print(f"{avg:.20f}")

if __name__ == "__main__":
    solve()
```
