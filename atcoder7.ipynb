```python
N,M = map(int, input().split())
A = list(tuple(map(int, input().split())) for i in range(N))
for j in range(1,M+1):
    count = 0
    total = 0
    for i in range(N):
        if A[i][0] == j:
            total += A[i][1]
            count += 1
    b = total / count
    B = "{:.20f}".format(b)
    print(B)
```
