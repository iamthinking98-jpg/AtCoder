```python
import sys
input = sys.stdin.readline

N = int(input())
A = list(map(int, input().split()))
    
count = 0

for i in range(N-1):
    current_sum = A[i]
    for j in range(i+1, N):
        judge = True
        current_sum += A[j]
        for k in range(i, j + 1):
            if current_sum % A[k] == 0:
                judge = False
                break
            
        if judge:
            count += 1
                
print(count)
```
