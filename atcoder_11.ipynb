```python
N = int(input())
A = list(map(int, input().split()))
T = True
start = 0
while T:
    if start < 4 and len(A) >3:
        for i in range(len(A)-3):
            if A[i] == A[i+1] == A[i+2] == A[i+3]:
                del A[i:i+4]
                start = i
                T = True
                break
            else:
                T = False
    elif len(A) >3:
        for i in range(start-3,len(A)-3):
            if A[i] == A[i+1] == A[i+2] == A[i+3]:
                del A[i:i+4]
                start = i
                T = True
                break
            else:
                T = False
    else:
        T = False
ansewr = len(A)
print(ansewr) 
```
