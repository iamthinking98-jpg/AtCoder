```python
N = int(input())
A = list(map(int, input().split()))

start = 0

while True:
    is_deleted = False
    search_start = max(0, start - 3)
    for i in range(search_start, len(A) - 3):
        if A[i] == A[i+1] == A[i+2] == A[i+3]:
            del A[i:i+4]
            start = i
            is_deleted = True
            break
    
    if is_deleted == False:
        break

print(len(A))
```
