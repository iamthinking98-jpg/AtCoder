```python
N,M = map(int, input().split())
def make_boxs(N,M):
    s_box = ["a","b","c","d","e","f","g","h","i","j"]
    b_box = list()
    for n in range(N):
        s_box[n] = list(input())
        b_box.append(s_box[n])
    return b_box
b_box = make_boxs(N,M)
total_box = list()
def make_total_box(b_box,N,M,limit):
    for m in range(limit,M + limit):
        t_box = b_box[m]
        for num in range(limit,M + limit):
            sub_box = list()
            sub_box.append(t_box[num])
    sub_box = tuple(sub_box)
    total_box.append(sub_box)
    return total_box
limit = 0
while limit != M:
    total_box = make_total_box(b_box,N,M)
    limit += 1
count = 0
for mum in range(len(total_box)-1):
    if total_box[mum] == total_box[mum+1]:
        count += 1
print(count)
```
