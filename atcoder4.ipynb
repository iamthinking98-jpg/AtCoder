```python
X = int(input())
N = int(input())
list_W = list(map(int, input().split()))
Q = int(input())
list_P = [int(input()) for q in range(Q)]
did_P = set()
for i in range(Q):
    s = list_P[i] - 1
    if s in did_P:
        X -= list_W[s]
        did_P.remove(s)
        print(X)
    else:
        X += list_W[s]
        did_P.add(s)
        print(X)
```
